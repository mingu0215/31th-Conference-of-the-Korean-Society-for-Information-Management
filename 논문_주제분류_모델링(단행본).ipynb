```python
def filter_exact_match(row, 중분류_목록):
    # 주제분류 컬럼의 값을 중분류 목록과 정확히 일치하는 경우만 남기고 중복 제거
    filtered = [sub for sub in row.split(', ') if sub in 중분류_목록]
    return ', '.join(sorted(set(filtered), key=filtered.index))

# 주제분류 칼럼에서 결측값이 있는 행을 삭제
df = df.dropna(subset=['주제분류'])

# '주제분류' 컬럼의 값이 중분류 목록에 정확히 일치하는 경우만 남기고 중복 제거
df['주제분류'] = df['주제분류'].apply(lambda x: filter_exact_match(x, 중분류_목록))

# 중분류 목록에 포함된 단어가 없는 행 제거
df = df[df['주제분류'] != '']
```

```python
def get_last_word(text):
    if isinstance(text, str):  # 문자열인지 확인
        words = text.split()  # 공백으로 단어 분리
        return words[-1] if words else text  # 마지막 단어 반환
    return text  # 문자열이 아닌 경우 원래 값 반환

df['주제분류'] = df['주제분류'].apply(get_last_word)

df['주제분류'].value_counts()
```

#### 1. Text Cleaning : 문자열 패턴 감지 및 조정(특수문자 제거)

```python
import pandas as pd
from tqdm import tqdm
import time
import re

# 데이터프레임 각 열에 대해 텍스트 정제
for column in columns:
    for i in tqdm(range(len(df))):

        # 특정 열에서 텍스트를 가져옴
        text = df.at[i, column]

        # 값이 0이 아닌 경우에만 정제 수행
        if text != 0:
            # 완성형 한글, 영어, 숫자를 제외한 모든 문자를 공백으로 대체
            cleaned_text = re.sub('[^가-힣a-zA-Z0-9]', ' ', text)

            # 띄어쓰기(공백)가 반복될 경우 1개로 축소
            cleaned_text = re.sub('[ ]{2,}', ' ', cleaned_text)

            # 결과를 다시 데이터프레임에 저장
            df.at[i, column] = cleaned_text

# 결과 확인
df.head()
```

```python
from konlpy.tag import Mecab
import nltk

df_together = df.copy()

# nltk punkt 데이터 다운로드
nltk.download('punkt')

# Mecab 형태소 분석기 초기화
mecab = Mecab()

# 데이터프레임 각 열에 대해 텍스트 토큰화 적용
for column in columns:
    for i in range(len(df_together)):
        text = df_together.iloc[i][column]

        # 값이 0이 아닌 경우에만 전처리 수행
        if text != 0:
            # nltk를 사용하여 토큰화
            nltk_tokens = nltk.word_tokenize(text)
            nltk_tokenized_text = ' '.join(nltk_tokens)

            # Mecab을 사용하여 추가 형태소 분석
            mecab_tokens = mecab.morphs(nltk_tokenized_text)

            # 결과를 기존 열에 덮어쓰기
            df_together.at[i, column] = ','.join(mecab_tokens)

# 결과 확인
df_together.head()
```

#### 3. 불용어 제거

```python
# 불용어 CSV 파일 불러오기
stopwords_csv = pd.read_csv("/content/drive/MyDrive/포트폴리오/정보관리학회 논문/stopwords.csv", header=None)
stopwords = set(stopwords_csv[0].tolist())  # 첫 번째 열을 리스트로 변환 후 세트로 변환
stopwords
```

```python
# 각 열에 대해 불용어 제거 수행
for column in tqdm(columns, desc='Processing columns'):
    for i in tqdm(range(len(df_together)), desc=f'Processing rows for column {column}', leave=False):
        if df_together.at[i, column] != 0:
            tokens = df_together.at[i, column].split(',')  # 콤마로 구분된 토큰을 다시 리스트로 변환
            filtered_tokens = [token for token in tokens if token not in stopwords]
            df_together.at[i, column] = ','.join(filtered_tokens)

# 결과 확인
df_together.head()
```

#### 4. 품사 태깅
- 명사(NNG), 고유명사(NNP)만 추출

```python
import pandas as pd
from konlpy.tag import Mecab

# Mecab 객체 초기화
mecab = Mecab()

# 명사와 고유명사만 추출하는 함수
def extract_nouns(tokens):
    return [f"{word}/{tag}" for word, tag in tokens if tag in ['NNG', 'NNP']]

# 데이터프레임 각 열에 대해 텍스트 토큰화 및 품사 태깅 적용
for column in columns:
    for i in range(len(df_together)):
        text = df_together.iloc[i][column]

        # 값이 0이 아닌 경우에만 전처리 수행
        if text != 0:
            # Mecab을 사용하여 품사 태깅
            tokens = mecab.pos(text)

            # 명사와 고유명사만 추출
            filtered_tokens = extract_nouns(tokens)

            # 태깅된 토큰을 문자열로 변환하여 저장
            tagged_tokens_str = ','.join(filtered_tokens)
            df_together.at[i, column] = tagged_tokens_str

# 결과 확인
df_together.head()
```

#### FastText 텍스트 임베딩

```python
import pandas as pd
from gensim.models import FastText

# FastText 모델 학습 함수
def train_fasttext(sentences, vector_size=100, window=3, min_count=1, workers=4):
    model = FastText(vector_size=vector_size, window=window, min_count=min_count, workers=workers)
    model.build_vocab(sentences)  # 단어 사전 생성
    model.train(sentences, total_examples=model.corpus_count, epochs=model.epochs)  # 모델 학습
    return model

# 문장 벡터화 함수
def vectorize_sentence(sentence, model, vector_size):
    words = [word.split('/')[0] for word in sentence.split()]  # 단어만 사용
    word_vecs = [model.wv[word] for word in words if word in model.wv]
    if len(word_vecs) == 0:
        return np.zeros(vector_size)
    else:
        return np.mean(word_vecs, axis=0)
```
